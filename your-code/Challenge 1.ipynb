```python
import pandas as pd
import numpy as np
import requests
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 100)
```

```python
repos = pd.read_json('../repos.json', orient='records')
repos
```

```python
forks = repos[repos['name'].str.contains('lab-')][['name', 'forks_url']]
forks
```

```python
languages = []
header = {'Authorization': 'token 4ba61ee862f283eb1226e398911deb54363339a4'} 
for url in forks['forks_url']:
    response = requests.get(url, headers=header)
    df = pd.DataFrame(response.json())
    if 'language' in df.columns:
        languages.append([language for language in df['language'].unique()])
```

```python
languages = set([language for lst in languages for language in lst]) # Flatten the list and give me unique values.
languages
```
